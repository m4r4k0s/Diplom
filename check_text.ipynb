```python
import os
import pandas as pd
import numpy as np
import pickle
import warnings
```

```python
with open("Classes_KMeans_4.pkl", "rb") as file_to_read:
    classes_KM = pickle.load(file_to_read)
    
classes_KM_np = np.asarray(classes_KM)
classes_KM_np.shape
```

```python
with open("embedded_file_names.pkl", "rb") as file_to_read:
    text_mapping = pickle.load(file_to_read)
len(text_mapping)
```

```python
q = [0,0,0,0,0,0]
text_maping_new = []
text_classes = []
for i in range (classes_KM_np.shape[0]):
    if classes_KM_np[i] == 0 and q[0] <5:
        text_maping_new.append(text_mapping[i])
        text_classes.append(classes_KM_np[i])
        print('текст 1 темы : ' , text_mapping[i] )
        q[0] +=1
    if classes_KM_np[i] == 1 and q[1] <5:
        text_maping_new.append(text_mapping[i])
        text_classes.append(classes_KM_np[i])
        print('текст 2 темы : ' , text_mapping[i] )
        q[1] +=1
    if classes_KM_np[i] == 2 and q[2] <5:
        text_maping_new.append(text_mapping[i])
        text_classes.append(classes_KM_np[i])
        print('текст 3 темы : ' , text_mapping[i] )
        q[2] +=1
    if classes_KM_np[i] == 3 and q[3] <5:
        text_maping_new.append(text_mapping[i])
        text_classes.append(classes_KM_np[i])
        print('текст 4 темы : ' , text_mapping[i] )
        q[3] +=1
    if classes_KM_np[i] == 4 and q[4] <5:
        text_maping_new.append(text_mapping[i])
        text_classes.append(classes_KM_np[i])
        print('текст 5 темы : ' , text_mapping[i] )
        q[4] +=1
    if classes_KM_np[i] == 5 and q[5] <5:
        text_maping_new.append(text_mapping[i])
        text_classes.append(classes_KM_np[i])
        print('текст 6 темы : ' , text_mapping[i] )
        q[5] +=1
```

```python
text_maping_new
```

```python
TRANSCRIBE_PATH = "recognition"
i = 0
q = []
for folder in os.listdir(TRANSCRIBE_PATH):
    if folder.startswith("."):
        continue
    print("folder: " + folder)
    folder_path = TRANSCRIBE_PATH + '/' + folder
    for text_name in os.listdir(folder_path):
        if text_name.startswith(".") or text_name not in text_maping_new:
            continue
        text_path = folder_path + '/' + text_name
        f = open(text_path, 'r')
        words = f.read()
        if words:
            print('Тема номер : ', text_classes[i])
            i +=1
            print(words)
            q.append(words)
            print('\\\\\\\\\\\\\\\\\\\\\\\\\\')
                    
```
